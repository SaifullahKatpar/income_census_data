```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

import matplotlib.pyplot as plt
import seaborn as sb
```

# preprocessing

```python
train = pd.read_csv('adult.data.csv')
test = pd.read_csv('adult.test.csv')
```

```python
data = pd.concat([train,test])
```

```python
cat_data = data.select_dtypes(include=['object']).copy()
cat_data.info()
```

```python
data.describe()
```

```python
for col in cat_data.columns:
    data[col] = data[col].astype('category')
    data[col] = data[col].cat.codes
```

```python
train = data[:32561]
test = data[32560:]
```

```python
corr = train.corr()
abs(corr>=0.3)
```

```python
from sklearn.linear_model import LogisticRegression #logistic regression
from sklearn import metrics #accuracy measure
from sklearn import svm #support vector Machine
from sklearn.ensemble import RandomForestClassifier #Random Forest
from sklearn.neighbors import KNeighborsClassifier #KNN
from sklearn.naive_bayes import GaussianNB #Naive bayes
from sklearn.tree import DecisionTreeClassifier #Decision Tree
```

```python
features = ['age', 'education-num', 'relationship', 'race', 'sex', 'gain', 'loss', 'hpw']
model = LogisticRegression()
model.fit(train[features],train['target'])
preds1 = model.predict(test[features])
acc1 = metrics.accuracy_score(preds1,test['target'])
acc1*100
```

```python
model=svm.SVC(kernel='rbf',C=1,gamma=0.1)
model.fit(train[features],train['target'])
prediction1=model.predict(test[features])
```

```python
print('Accuracy for rbf SVM is ',metrics.accuracy_score(prediction1,test['target'])*100)
```

```python
model=svm.SVC(kernel='linear',C=0.1,gamma=0.1)
model.fit(train[features],train['target'])
prediction2=model.predict(test[features])
print('Accuracy for linear SVM is',metrics.accuracy_score(prediction2,test['target'])*100)
```
